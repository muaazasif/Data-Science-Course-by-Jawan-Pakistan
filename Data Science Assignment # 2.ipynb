# ADD COLUMNS

```python
speed_mean=data["Speed"].mean()
speed_mean
```

```python
def sp_mean(val):
    if val < speed_mean:
        return "Low Speed"
    elif val==speed_mean:
        return "Neutral Speed"
    else:
        return "High Speed"
```

# High_Low_Speed

```python
data["High_Low_Speed"]=data["Speed"].apply(sp_mean)
data
```

```python
hp_mean=data["HP"].mean()
hp_mean
```

```python
def HP(val):
    if val < hp_mean:
        return "Low HP"
    elif val==hp_mean:
        return "Neutral HP"
    else:
        return "High HP"
```

# High_Low_HP

```python
data["High_Low_HP"]=data["HP"].apply(HP)
data
```
